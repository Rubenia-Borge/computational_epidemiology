```python
# Random Walks 1

import random

Total_Times_We_Run=1000

n=100 # number of random steps to take
data_question_2 = open("data_question_2.csv","w") # open file for writing
data_question_2.write("Times,Steps, Step at Infection,Infection?, X1, Y1, X2, Y2, Probability, Manhattan Distance\n") # write headers to file
data_infected_2 = open("data_infected_2.csv","w")
data_infected_2.write("Times, Steps,X1, Y1, X2, Y2,Manhattan,Is_Infected\n")


for times in range(1, Total_Times_We_Run + 1):
    Px=0#x-position
    Py=0 #y-position
    Px2 = 10
    Py2 = 10
    Infected = 0
    step_at_infection=0
    Is_Infected = "No"
    S=0 #step counter
    probability = 0
    manhattan = 0

    while (S<n): #while we have not yet taken n steps
        manhattan = (abs(Px2 - Px) + abs(Py2 - Py))
        data_question_2.write(
            str(times) + "," + str(S) + "," + str(step_at_infection) + "," + Is_Infected + "," + str(Px) + "," + str(
                Py) + "," + str(Px2) + "," + str(Py2) + "," + str(probability) + "," + str(
                manhattan) + "\n")  # write the final position
        R=random.random()
        if R<0.25:
            Px-=1 #move left
            if Px < 0:
                Px = 1
            if (Px == Px2) and (Py == Py2):
                Is_Infected = "Yes"
                Infected = Infected + 1
                step_at_infection = S
                probability = Infected / 1000
                manhattan = (abs(Px2 - Px) + abs(Py2 - Py))
                data_infected_2.write(str(times)+","+str(S)+","+str(Px)+","+str(Py)+","+str(Px2)+","+str(Py2)+","+str(manhattan)+","+str(Is_Infected)+","+"\n")
                data_question_2.write(
                    str(times) + "," + str(S) + "," + str(step_at_infection) + "," + Is_Infected + "," + str(
                        Px) + "," + str(Py) + "," + str(Px2) + "," + str(Py2) + "," + str(probability) + "," + str(
                        manhattan) + "\n")  # write the final position of this run to file
            else:
                Is_Infected = "No"


        elif R<0.5:
            Px+=1 #move right
            if Px > 100:
                Px = 100
            if (Px == Px2) and (Py == Py2):
                Is_Infected = "Yes"
                Infected = Infected + 1
                step_at_infection = S
                probability = Infected / 1000
                manhattan = (abs(Px2 - Px) + abs(Py2 - Py))
                data_infected_2.write(
                    str(times) + "," + str(S) + "," + str(Px) + "," + str(Py) + "," + str(Px2) + "," + str(
                        Py2) + "," + str(manhattan) + "," + str(Is_Infected) + "," + "\n")
                data_question_2.write(
                    str(times) + "," + str(S) + "," + str(step_at_infection) + "," + Is_Infected + "," + str(
                        Px) + "," + str(Py) + "," + str(Px2) + "," + str(Py2) + "," + str(probability) + "," + str(
                        manhattan) + "\n")  # write the final position of this run to file
            else:
                Is_Infected = "No"

        elif R<0.75:
            Py-=1 #move down
            if Py < 0:
                Py = 1
            if (Px == Px2) and (Py == Py2):
                Is_Infected = "Yes"
                Infected = Infected + 1
                step_at_infection = S
                probability = Infected / 1000
                manhattan = (abs(Px2 - Px) + abs(Py2 - Py))
                data_infected_2.write(
                    str(times) + "," + str(S) + "," + str(Px) + "," + str(Py) + "," + str(Px2) + "," + str(
                        Py2) + "," + str(manhattan) + "," + str(Is_Infected) + "," + "\n")
                data_question_2.write(
                    str(times) + "," + str(S) + "," + str(step_at_infection) + "," + Is_Infected + "," + str(
                        Px) + "," + str(Py) + "," + str(Px2) + "," + str(Py2) + "," + str(probability) + "," + str(
                        manhattan) + "\n")  # write the final position of this run to file
            else:
                Is_Infected = "No"


        else:
            Py+=1 #move up
            if Py > 100:
                Py = 100

            if (Px == Px2) and (Py == Py2):
                Is_Infected = "Yes"
                Infected = Infected + 1
                step_at_infection = S
                probability = Infected / 1000
                manhattan = (abs(Px2 - Px) + abs(Py2 - Py))
                data_infected_2.write(
                    str(times) + "," + str(S) + "," + str(Px) + "," + str(Py) + "," + str(Px2) + "," + str(
                        Py2) + "," + str(manhattan) + "," + str(Is_Infected) + "," + "\n")
                data_question_2.write(
                    str(times) + "," + str(S) + "," + str(step_at_infection) + "," + Is_Infected + "," + str(
                        Px) + "," + str(Py) + "," + str(Px2) + "," + str(Py2) + "," + str(probability) + "," + str(
                        manhattan) + "\n")  # write the final position of this run to file
            else:
                Is_Infected = "No"



        R2 = random.random()
        if R2 < 0.25:
            Px2 -= 1  # move left
            if Px2 < 0:
                Px2 = 1
            if (Px == Px2) and (Py == Py2):
                Is_Infected = "Yes"
                Infected = Infected + 1
                step_at_infection = S
                probability = Infected / 1000
                manhattan = (abs(Px2 - Px) + abs(Py2 - Py))
                data_infected_2.write(
                    str(times) + "," + str(S) + "," + str(Px) + "," + str(Py) + "," + str(Px2) + "," + str(
                        Py2) + "," + str(manhattan) + "," + str(Is_Infected) + "," + "\n")
                data_question_2.write(
                    str(times) + "," + str(S) + "," + str(step_at_infection) + "," + Is_Infected + "," + str(
                        Px) + "," + str(Py) + "," + str(Px2) + "," + str(Py2) + "," + str(probability) + "," + str(
                        manhattan) + "\n")  # write the final position of this run to file
            else:
                Is_Infected = "No"


        elif R2 < 0.5:
            Px2 += 1  # move right
            if Px2 > 100:
                Px2 = 100
            if (Px == Px2) and (Py == Py2):
                Is_Infected = "Yes"
                Infected = Infected + 1
                step_at_infection = S
                probability = Infected / 1000
                manhattan = (abs(Px2 - Px) + abs(Py2 - Py))
                data_infected_2.write(
                    str(times) + "," + str(S) + "," + str(Px) + "," + str(Py) + "," + str(Px2) + "," + str(
                        Py2) + "," + str(manhattan) + "," + str(Is_Infected) + "," + "\n")
                data_question_2.write(
                    str(times) + "," + str(S) + "," + str(step_at_infection) + "," + Is_Infected + "," + str(
                        Px) + "," + str(Py) + "," + str(Px2) + "," + str(Py2) + "," + str(probability) + "," + str(
                        manhattan) + "\n")  # write the final position of this run to file
            else:
                Is_Infected = "No"


        elif R2 < 0.75:
            Py2 -= 1  # move down
            if Py2 < 0:
                Py2 = 1
            if (Px == Px2) and (Py == Py2):
                Is_Infected = "Yes"
                Infected = Infected + 1
                step_at_infection = S
                probability = Infected / 1000
                manhattan = (abs(Px2 - Px) + abs(Py2 - Py))
                data_infected_2.write(
                    str(times) + "," + str(S) + "," + str(Px) + "," + str(Py) + "," + str(Px2) + "," + str(
                        Py2) + "," + str(manhattan) + "," + str(Is_Infected) + "," + "\n")
                data_question_2.write(
                    str(times) + "," + str(S) + "," + str(step_at_infection) + "," + Is_Infected + "," + str(
                        Px) + "," + str(Py) + "," + str(Px2) + "," + str(Py2) + "," + str(probability) + "," + str(
                        manhattan) + "\n")  # write the final position of this run to file
            else:
                Is_Infected = "No"

        else:
            Py2 += 1  # move up
            if Py2 > 100:
                Py2 = 100
            if (Px == Px2) and (Py == Py2):
                Is_Infected = "Yes"
                Infected = Infected + 1
                step_at_infection = S
                probability = Infected / 1000
                manhattan = (abs(Px2 - Px) + abs(Py2 - Py))
                data_infected_2.write(
                    str(times) + "," + str(S) + "," + str(Px) + "," + str(Py) + "," + str(Px2) + "," + str(
                        Py2) + "," + str(manhattan) + "," + str(Is_Infected) + "," + "\n")
                data_question_2.write(
                    str(times) + "," + str(S) + "," + str(step_at_infection) + "," + Is_Infected + "," + str(
                        Px) + "," + str(Py) + "," + str(Px2) + "," + str(Py2) + "," + str(probability) + "," + str(
                        manhattan) + "\n")  # write the final position of this run to file
            else:
                Is_Infected = "No"

        S+=1 #update step counter

print('Program finished. Please check csv files generated for results.')
data_question_2.close()
data_infected_2.close()
```
